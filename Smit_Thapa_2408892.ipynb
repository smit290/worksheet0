```python
#Task - 3 - Outlier Detection:

#1 Calculate Q1, Q3, and IQR for 'Score'


Q1_score = dataset['score'].quantile(0.25)
Q3_score = dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

#2 Calculate Q1, Q3, and IQR for 'Log GDP per capita'
Q1_gdp = dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers for 'Score' and 'Log GDP per capita'
dataset['Score Outlier'] = (dataset['score'] < lower_bound_score) | (dataset['score'] > upper_bound_score)
dataset['GDP Outlier'] = (dataset['Log GDP per capita'] < lower_bound_gdp) | (dataset['Log GDP per capita'] > upper_bound_gdp)

# Display the rows that are outliers
outliers = dataset[(dataset['Score Outlier']) | (dataset['GDP Outlier'])]
print(outliers[['Country name', 'score', 'Log GDP per capita', 'Score Outlier', 'GDP Outlier']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
dataset = pd.read_csv('/content/south_asia_filtered.csv')

# Print column names to check the correct column name
print(dataset.columns)

# Replace 'Score' with the actual column name from the dataset
# Example: If the column name is 'score' instead of 'Score'
score_col = 'score'  # Change this based on your dataset
gdp_col = 'Log GDP per capita'

# Calculate Q1, Q3, and IQR for 'score'
Q1_score = dataset[score_col].quantile(0.25)
Q3_score = dataset[score_col].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Calculate lower and upper bounds for 'score'
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Calculate Q1, Q3, and IQR for 'Log GDP per capita'
Q1_gdp = dataset[gdp_col].quantile(0.25)
Q3_gdp = dataset[gdp_col].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Calculate lower and upper bounds for 'Log GDP per capita'
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers for 'score' and 'Log GDP per capita'
dataset['Score Outlier'] = (dataset[score_col] < lower_bound_score) | (dataset[score_col] > upper_bound_score)
dataset['GDP Outlier'] = (dataset[gdp_col] < lower_bound_gdp) | (dataset[gdp_col] > upper_bound_gdp)

# Combine outlier columns into a single 'Outlier' column
dataset['Outlier'] = dataset['Score Outlier'] | dataset['GDP Outlier']

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=gdp_col,
    y=score_col,
    data=dataset,
    hue='Outlier',  # Differentiate outliers
    palette={False: 'blue', True: 'red'},  # Blue for normal, red for outliers
    s=100  # Marker size
)

# Add labels and title
plt.title('Scatter Plot of GDP per Capita vs Score (Outliers Highlighted)', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.legend(title='Outlier', labels=['Not Outlier', 'Outlier'])
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
#Task - 5 - Gap Analysis:
#1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asia dataset
dataset = pd.read_csv('/content/south_asia_filtered.csv')

# Add the GDP-Score Gap column
dataset['GDP-Score Gap'] = dataset['Log GDP per capita'] - dataset['score']

# Display the updated dataset
print(dataset[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])
```

```python
#2
# Rank countries by GDP-Score Gap in ascending and descending order
ascending_gap = dataset.sort_values(by='GDP-Score Gap', ascending=True)
descending_gap = dataset.sort_values(by='GDP-Score Gap', ascending=False)

# Display rankings
print("\nTop 3 Countries with the Largest Negative GDP-Score Gap:")
print(ascending_gap[['Country name', 'GDP-Score Gap']].head(3))

print("\nTop 3 Countries with the Largest Positive GDP-Score Gap:")
print(descending_gap[['Country name', 'GDP-Score Gap']].head(3))
```

```python
#3
# Select top 3 countries for both positive and negative gaps
top_positive = descending_gap[['Country name', 'GDP-Score Gap']].head(3)
top_negative = ascending_gap[['Country name', 'GDP-Score Gap']].head(3)

# Combine the results for plotting
top_countries = pd.concat([top_positive, top_negative])

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    x='GDP-Score Gap',
    y='Country name',
    data=top_countries,
    palette='coolwarm'
)

# Add title and labels
plt.title('Top 3 Positive and Negative GDP-Score Gaps', fontsize=14)
plt.xlabel('GDP-Score Gap', fontsize=12)
plt.ylabel('Country Name', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#3. Metric Comparisons:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets for South Asia and Middle East
south_asia_df = pd.read_csv('south_asia_filtered.csv')
middle_east_df = pd.read_csv('middle_east_filtered.csv')

# Select key metrics for comparison
key_metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

# Calculate mean values for each region
south_asia_means = south_asia_df[key_metrics].mean()
middle_east_means = middle_east_df[key_metrics].mean()

# Combine the mean values into a single DataFrame
comparison_df = pd.DataFrame({
    'Metric': key_metrics,
    'South Asia': south_asia_means.values,
    'Middle East': middle_east_means.values
})

# Reshape for grouped bar chart
comparison_df = comparison_df.melt(id_vars='Metric', var_name='Region', value_name='Value')

# Plot grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    data=comparison_df,
    x='Metric',
    y='Value',
    hue='Region',
    palette='viridis'
)

# Add labels and title
plt.title('Comparison of Key Metrics Between South Asia and Middle East')
plt.xlabel('Metric')
plt.ylabel('Mean Value')
plt.legend(title='Region')
plt.xticks(rotation=15)
plt.tight_layout()

# Show the plot
plt.show()

# Identify the largest disparity
disparities = abs(south_asia_means - middle_east_means)
largest_disparity_metric = disparities.idxmax()
largest_disparity_value = disparities.max()

# Print the metric with the largest disparity
print(f"The metric with the largest disparity is '{largest_disparity_metric}' with a difference of {largest_disparity_value:.2f}.")
```

```python
#Outlier Detection:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets for South Asia and Middle East
south_asia_df = pd.read_csv('south_asia_filtered.csv')
middle_east_df = pd.read_csv('middle_east_filtered.csv')

# Function to detect outliers using the IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers based on 'score' and 'GDP per capita' for both regions
south_asia_score_outliers = detect_outliers(south_asia_df, 'score')
south_asia_gdp_outliers = detect_outliers(south_asia_df, 'Log GDP per capita')

middle_east_score_outliers = detect_outliers(middle_east_df, 'score')
middle_east_gdp_outliers = detect_outliers(middle_east_df, 'Log GDP per capita')

# Plotting
plt.figure(figsize=(14, 7))

# South Asia - Score vs GDP per Capita
plt.subplot(1, 2, 1)
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Non-Outliers', color='blue')
sns.scatterplot(x=south_asia_score_outliers['Log GDP per capita'], y=south_asia_score_outliers['score'], color='red', label='Score Outliers')
sns.scatterplot(x=south_asia_gdp_outliers['Log GDP per capita'], y=south_asia_gdp_outliers['score'], color='green', label='GDP Outliers')
plt.title('South Asia: Outliers based on Score and GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

# Middle East - Score vs GDP per Capita
plt.subplot(1, 2, 2)
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_df, label='Non-Outliers', color='blue')
sns.scatterplot(x=middle_east_score_outliers['Log GDP per capita'], y=middle_east_score_outliers['score'], color='red', label='Score Outliers')
sns.scatterplot(x=middle_east_gdp_outliers['Log GDP per capita'], y=middle_east_gdp_outliers['score'], color='green', label='GDP Outliers')
plt.title('Middle East: Outliers based on Score and GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Visualization:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets for South Asia and Middle East
south_asia_df = pd.read_csv('south_asia_filtered.csv')
middle_east_df = pd.read_csv('middle_east_filtered.csv')

# Add a new column to differentiate the regions for plotting
south_asia_df['Region'] = 'South Asia'
middle_east_df['Region'] = 'Middle East'

# Combine the two datasets into one for easier plotting
combined_df = pd.concat([south_asia_df[['score', 'Region']], middle_east_df[['score', 'Region']]])

# Create the boxplot comparing the distribution of Score between South Asia and Middle East
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='score', data=combined_df, palette='Set2')

# Add titles and labels
plt.title('Comparison of Happiness Scores: South Asia vs Middle East')
plt.xlabel('Region')
plt.ylabel('Happiness Score')

# Show the plot
plt.show()
```

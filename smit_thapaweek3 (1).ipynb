exercise 1.draw the line represented by the function y=x

```python
import matplotlib.pyplot as plt

# Generating data points for the line y=x
x = []
y = []
for i in range(-10, 11):
    x.append(i)
    y.append(i)

# Plotting the line
plt.plot(x, y, label='y = x')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot of y = x')

# Displaying the legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```

q.no2. Draw a line in a diagram that joins the position (1, 3), (2, 8), (6, 1) and (8, 10).

```python
x = [1, 2, 6, 8]
y = [3, 8, 1, 10]
plt.plot(x, y, 'bo')  # 'bo' specifies blue dots
plt.plot(x, y, 'r-')  #'r-' specifies a red solid line
plt.title("connection")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.show()
```

q.no.3.Plot 20 values of x and y from the following functions

```python
import matplotlib.pyplot as plt
import numpy as np

def function_1(x):
    return 2 * x + 5

def function_2(x):
    return x ** 2

x_values = np.linspace(0, 10, 20)

y_values_1 = function_1(x_values)
y_values_2 = function_2(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values_1, label='2x + 5')
plt.title('Plot of Function 1: 2x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values_2, label='x^2')
plt.title('Plot of Function 2: x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
